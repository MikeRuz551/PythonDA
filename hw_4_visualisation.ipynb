## 3. Barpot

Постройте Bar-график суммарного количества прокатов велосипедов по месяцам за каждый год одновременно. (будет 24 столбика)

**Ответьте на вопрос:** почему в предыдущем графике была такая большая корреляция между количеством прокатов и годом?

Ответ: корреляция между количеством проката и годом не показательна, нет функциональной зависимости. Большая корреляция, потому что в отдельно взятому году большое кол-во проката.

```python
plt.figure (figsize=(8, 6), dpi=80)
months = []
cnt = []
for i in (0, 1):
  for j in range(1, 13):
    months.append(f'{j}.{i + 2011}')
    cnt.append(sum(df[(df['mnth'] == j) & (df['yr'] == i)]['cnt']))
plt.xticks(rotation=60)
plt.bar(months, cnt)
```
