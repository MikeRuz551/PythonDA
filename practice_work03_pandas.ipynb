11) Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак marital-status)? Женатыми считаем тех, у кого marital-status начинается с Married (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.

```python
mask = (df['marital.status'] == 'Married-civ-spouse') |  (df['marital.status'] == 'Married-spouse-absent') | (df['marital.status'] == 'Married-AF-spouse')

married = df[mask & (df['sex'] == 'Male')]['income'] == '>50K'
not_mar = df[~(mask & (df['sex'] == 'Male'))]['income'] == '>50K'

print(f'''Доля хорошо зарабатывающих среди женатых = {married.mean():.2f}
Доля хорошо зарабатывающих среди холостых = {not_mar.mean():.2f}''')
```

12) Посчитайте среднее время работы (hours-per-week) зарабатывающих мало и много (salary) для каждой страны (native-country).

```python
agg_func_math = {'hours.per.week': 'mean'}
s1 = df[df['income'] == '<=50K'].groupby('native.country').agg(agg_func_math)
s2 = df[df['income'] == '>50K'].groupby('native.country').agg(agg_func_math)
s1 = s1.rename(columns={'hours.per.week': 'small salary'})
s1['big salary'] = s2
s1
```

15) Создайте случайную колонку - magic_salary, которую нужно будет вычислить следующим образом: если зарплата небольшая (<50K), тогда случайно выберите число из диапазона [0,50]. Если зарплата выше 50K тогда из диапазона [51, 200]. Посчитайте среднюю зарплату в час для групп людей с одни уровнем образования на основе нашей случайной колонки magic_salary

```python
df['magic_salary'] = df['income'].where(df['income'] != '<=50K', np.random.uniform(0, 50, len(df)))
df['magic_salary'][df['magic_salary'] == '>50K'] = df['income'].where(df['income'] != '>50K', np.random.uniform(51, 200, len(df)))
agg_func_math = {
    'magic_salary': 'mean',
    'hours.per.week': 'mean'}
data = df.groupby('education').agg(agg_func_math)
(data['magic_salary'] / data['hours.per.week']).round(2)
```
